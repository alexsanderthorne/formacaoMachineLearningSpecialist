```python
import cv2
from google.colab.patches import cv2_imshow

# Carregar a imagem
imagem = cv2.imread('/content/WhatsApp Image 2024-01-26 at 15.34.38(2).jpeg')

# Transformar a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Binarizar a imagem
_, imagem_binarizada = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)

# Exibir as imagens
cv2_imshow(imagem)
cv2_imshow(imagem_cinza)
cv2_imshow(imagem_binarizada)
```

```python
from PIL import Image

def rgb_to_grayscale(image):
    width, height = image.size
    grayscale_image = Image.new('L', (width, height))

    for x in range(width):
        for y in range(height):
            r, g, b = image.getpixel((x, y))
            gray_value = int(0.3 * r + 0.59 * g + 0.11 * b)
            grayscale_image.putpixel((x, y), gray_value)

    return grayscale_image

def binarize_image(image, threshold=127):
    width, height = image.size
    binarized_image = Image.new('L', (width, height))

    for x in range(width):
        for y in range(height):
            pixel_value = image.getpixel((x, y))
            binarized_value = 0 if pixel_value < threshold else 255
            binarized_image.putpixel((x, y), binarized_value)

    return binarized_image

# Carregar a imagem
imagem = Image.open('/content/WhatsApp Image 2024-01-26 at 15.34.38(2).jpeg')

# Converter para tons de cinza
imagem_cinza = rgb_to_grayscale(imagem)

# Binarizar a imagem
imagem_binarizada = binarize_image(imagem_cinza)

# Salvar as imagens
imagem_cinza.save('/content/imagem_cinza.jpg')
imagem_binarizada.save('/content/imagem_binarizada.jpg')
```
